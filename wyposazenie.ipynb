```python
import random
import numpy as np
import mysql.connector
import pandas as pd
```

```python
data = pd.read_excel("wyposazenie.xlsx")
data
```

```python
def price(index):
    if index < 8:
        return round(random.uniform(8000,15000), 2)
    elif 8 <= index <= 22:
        return round(random.uniform(300,700), 2)
    elif 22 < index <= 43:
        return round(random.uniform(100,300), 2)
    else:
        return round(random.uniform(7, 15), 2) 

indexes = {equipment: index for index, equipment in enumerate(data.iloc[:, 0])}
prices = {equipment: price(index) for equipment, index in indexes.items()}

data['Cena'] = data.iloc[:, 0].map(prices)
data
    
```

```python
def avaiable(index):
    if index < 8:
        return random.randint(0,6)
    elif 8 <= index <= 22:
        return random.randint(0,11)
    elif 22 < index <= 43:
        return random.randint(0,16)
    else:
        return random.randint(0,21)
    
indexes = {equipment: index for index, equipment in enumerate(data.iloc[:, 0])}
avaiable_equipment = {equipment: avaiable(index) for equipment, index in indexes.items()}

data['Ilość_dostępnych'] = data.iloc[:, 0].map(avaiable_equipment)
data
    
```
